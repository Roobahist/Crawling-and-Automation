```python
def tolidforoshintimeframe(instrument,sector_id,sector,symbol,name,exchange,timeframe,par,cookies):
    r = requests.get('http://www.bourseview.com/api/v2/exchanges/IRTSENO/stocks/'+ instrument +'/productions',params=par,cookies=cookies)
#     print(r.status_code)
    j = r.json()
    
    items = j['items']
    href = j['href']
    instrument = href.split('/')[8]
    
    data = pd.DataFrame(columns=['symbol', 'name','instrument', 'exchange', 'sector', 'sector_id', 'fiscalMonth', 'fiscalYear','timeframe','periodEndingDate', 'auditing', 'company_key', 'productKey', 'productName', 'productionItemKey','productionItemName', 'publishTime', 'scenario', 'valueGrowthRate', 'valueGrowthRateYoy', 'valueOrigin','unitKey','unitName','valueVertical', 'statementKey', 'statementType','htmlUrl','attachmentUrl','display_order'])
    for i in items:
        record = dict()
        record['timeframe'] = timeframe
        record['instrument'] = instrument
        record['symbol'] = symbol
        record['sector_id'] = sector_id
        record['sector'] = sector
        record['name'] = name
        record['exchange'] = exchange
        record['statementType'] = i['statementType']
        record['fiscalMonth'] = i['fiscalMonth']
        record['fiscalYear'] = i['fiscalYear']
        record['periodEndingDate'] = i['periodEndingDate']
        record['scenario'] = i['scenario']
        record['auditing'] = i['auditing']
        record['htmlUrl'] = i['htmlUrl']
        record['attachmentUrl'] = i['attachmentUrl']
        record['company_key'] = i['company_key']
        record['publishTime'] = i['publishTime']
        record['statementKey'] = i['statementKey']
        productionItems = i['productionItems']
        for j in productionItems:
            record['productionItemKey'] = j['productionItemKey']
            record['productionItemName'] = j['productionItemName']
            record['productKey'] = j['productKey']
            record['productName'] = j['productName']
            record['valueOrigin'] = j['valueOrigin']
            record['valueVertical'] = j['valueVertical']
            record['valueGrowthRate'] = j['valueGrowthRate']
            record['valueGrowthRateYoy'] = j['valueGrowthRateYoy']
            record['unitKey'] = j['unitKey']
            record['unitName'] = j['unitName']
            record['display_order'] = j['display_order']
            data = data.append(record,ignore_index=True)
            
    return j , data
```
